```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/cleaned_dataset.csv")
sns.set(style="whitegrid")
plt.figure(figsize=(8,5))
sns.histplot(df['Dining_Rating'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Dining Ratings')
plt.xlabel('Dining Rating')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/cleaned_dataset.csv")
sns.set(style="whitegrid")
plt.figure(figsize=(12,6))
avg_rating_per_cuisine = df.groupby('Cuisine')['Dining_Rating'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=avg_rating_per_cuisine.values, y=avg_rating_per_cuisine.index, palette='viridis')
plt.title('Top 10 Cuisines by Average Dining Rating')
plt.xlabel('Average Dining Rating')
plt.ylabel('Cuisine')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/cleaned_dataset.csv")
sns.set(style="whitegrid")
plt.figure(figsize=(8,5))
sns.scatterplot(x='Prices', y='Average_Rating', data=df, alpha=0.5)
plt.title('Prices vs Average Rating')
plt.xlabel('Prices')
plt.ylabel('Average Rating')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/cleaned_dataset.csv")
sns.set(style="whitegrid")
plt.figure(figsize=(10,8))
numeric_df = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_df.corr(), annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
